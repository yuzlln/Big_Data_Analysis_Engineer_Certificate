67. Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다. 해당경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하라

```python
import datetime

def fix_centry(x) :
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_centry)

ans = df.copy()
ans.head()
```

69. weekday컬럼을 만들고 요일별로 매핑하라 ( 월요일 : 0 ~ 일요일 : 6)

70. weekday컬럼을 기준으로 주말이면 1 평일이면 0의 값을 가지는 WeekCheck 컬럼을 만들어라

```python
df['WeekCheck'] = df['weekday'].map(lambda x : 1 if x in [5, 6] else 0).to_frame()

ans = df[['weekday', 'WeekCheck']]
ans.head()
```

76. 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다

```python
# 데이터 로드
DataURL = 'https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv'
df = pd.read_csv(DataURL)

print(df.shape)
df.head(2)
```

```python
# datetime 형태로 변환
import datetime

def change_date(x) : 
    date = x.split(':')[0] # 년월일
    hour = x.split(':')[1] # 시

    if hour == '24' : # 24시인 경우 하루 추가
        hour = '00:00:00'
        FinalDate = pd.to_datetime(date + ' ' + hour) + datetime.timedelta(days=1)
    
    else :
        hour = hour + ':00:00'
        FinalDate = pd.to_datetime(date + ' ' + hour)
    
    return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)

ans = df.copy()
ans.head(2)
```

77. 일자별 영어요일 이름을 dayName 컬럼에 저장하라

```python
df['dayName'] = df['(년-월-일:시)'].dt.day_name()

ans = df[['(년-월-일:시)', 'dayName']]
ans.head()
```

78. 일자별 각 PM10등급의 빈도수를 파악하라

```python
# 방법2
ans2 = df.groupby([df['(년-월-일:시)'].dt.day, 'PM10등급'], as_index=False).size()

ans2 = ans2.pivot(index='(년-월-일:시)'
                  , columns='PM10등급'
                  , values='size').fillna(0)

ans2.head()
```

79. 시간이 연속적으로 존재하며 결측치가 없는지 확인하라

```python
# 1차 차분 (현재 값과 이전 값의 차이)
# 시간을 차분했을 경우 첫 값은 nan, 이후 모든 차분값이 동일하면 연속이라 판단

check = len(df['(년-월-일:시)'].diff().unique())

if check == 2 :
    ans = '시간이 연속적으로 존재'
else : 
    ans = '시간이 불연속적으로 존재'

ans
```

80. 오전 10시와 오후 10시(22시)의 PM10의 평균값을 각각 구하여라

```python
ans = df[(df['(년-월-일:시)'].dt.hour == 10) | (df['(년-월-일:시)'].dt.hour == 22)]
ans = ans.groupby(df['(년-월-일:시)'].dt.hour)['PM10'].mean().to_frame()
ans
```

82. 데이터를 주단위로 뽑아서 최소,최대, 평균, 표준표차를 구하여라

```python
# resample : 데이터프레임 df의 인덱스가 DatetimeIndex일 때 사용 가능
# 기본적으로 한 주는 일요일로 종료되며, 한 주에 속한 데이터를 하나의 그룹으로 묶음
ans = df.resample('W')['PM10', 'PM2.5', '오존', '이산화질소'].agg(['min', 'max', 'mean', 'std'])
ans.head(3)
```
