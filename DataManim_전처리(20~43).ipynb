# 2. Filtering & Sorting

```python
import pandas as pd
import numpy as np
```

20. 데이터를 로드하라

```python
DataURL = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df = pd.read_csv(DataURL)

ans = type(df)
ans
```

21. quantity컬럼 값이 3인 데이터를 추출하여 첫 5행을 출력하라

```python
ans = df[df['quantity'] == 3].head()
ans
```

22. quantity컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행을 출력하라

```python
# drop=True : 기존의 인덱스를 새로운 열로 추가하지 않고 완전히 제거
ans = df.loc[df['quantity'] == 3].reset_index(drop=True).head()
ans
```

23. quantity , item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하라

```python
ans = df[['quantity', 'item_price']]
ans
```

24. item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하라

```python
df['new_price'] = df['item_price'].str.replace('$', '').astype('float')

ans = df[['item_price', 'new_price']].head()
ans
```

25. new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하여라

```python
ans = len(df[df['new_price'] <= 5])
ans
```

26. item_name명이 Chicken Salad Bowl 인 데이터 프레임을 추출하라고 index 값을 초기화 하여라

```python
ans = df[df['item_name'] == 'Chicken Salad Bowl'].reset_index(drop=True)
ans.head()
```

27. new_price값이 9 이하이고 item_name 값이 Chicken Salad Bowl 인 데이터 프레임을 추출하라

```python
ans = df[(df['new_price'] <= 9) & (df['item_name'] == 'Chicken Salad Bowl')]
ans.head()
```

28. df의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화 하여라

```python
ans = df.sort_values('new_price').reset_index(drop=True)
ans.head()
```

29. df의 item_name 컬럼 값 중 Chips 포함하는 경우의 데이터를 출력하라

```python
ans = df[df['item_name'].str.contains('Chips')]
ans.head()
```

30. df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력하라

```python
ans = df.iloc[:, ::2]
ans.head()
```

30-1. df의 홀수번째 칼럼만을 포함하는 데이터프레임을 출력하라

```python
ans = df.iloc[:, 1::2]
ans.head()
```

31. df의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index를 초기화 하여라

```python
ans = df.sort_values('new_price', ascending=False).reset_index(drop=True)
ans.head()
```

32. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 인덱싱하라

```python
ans = df[(df['item_name'] == 'Steak Salad') | (df['item_name'] == 'Bowl')]
ans.head()
```

33. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라

```python
ans = df[(df['item_name'] == 'Steak Salad') | (df['item_name'] == 'Bowl')]
ans = ans.drop_duplicates(subset='item_name', keep='first')
ans
```

34. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남겨라

```python
ans = df[(df['item_name'] == 'Steak Salad') | (df['item_name'] == 'Bowl')]
ans = ans.drop_duplicates(subset='item_name', keep='last')
ans
```

35. df의 데이터 중 new_price값이 new_price값의 평균값 이상을 가지는 데이터들을 인덱싱하라

```python
mean = df['new_price'].mean()

ans = df[df['new_price'] >= mean]
ans.head()
```

36. df의 데이터 중 item_name의 값이 Izze 데이터를 Fizzy Lizzy로 수정하라

```python
df['item_name'] = df['item_name'].str.replace('Izze', 'Fizzy Lizzy')

ans = df
ans.head()
```

37. df의 데이터 중 choice_description 값이 NaN 인 데이터의 갯수를 구하여라

```python
ans = df['choice_description'].isna().sum()
ans
```

38. df의 데이터 중 choice_description 값이 NaN 인 데이터를 NoData 값으로 대체하라(loc 이용)

```python
df.loc[df['choice_description'].isna(), 'choice_description'] = 'NoData'

ans = df
ans.head()
```

39. df의 데이터 중 choice_description 값에 Black이 들어가는 경우를 인덱싱하라

```python
ans = df[df['choice_description'].str.contains('Black')]
ans.head()
```

40. df의 데이터 중 choice_description 값에 Vegetables 들어가지 않는 경우의 갯수를 출력하라

```python
ans = len(df[~df['choice_description'].str.contains('Vegetables')])
ans
```

41. df의 데이터 중 item_name 값이 N으로 시작하는 데이터를 모두 추출하라

```python
ans = df[df['item_name'].str.startswith('N')]
ans.head()
```

42. df의 데이터 중 item_name 값의 단어갯수가 15개 이상인 데이터를 인덱싱하라

```python
ans = df[df['item_name'].str.len() >= 15]
ans.head()
```

43. df의 데이터 중 new_price값이 lst에 해당하는 경우의 데이터 프레임을 구하고 그 갯수를 출력하라

```python
lst =[1.69, 2.39, 3.39, 4.45, 9.25, 10.98, 11.75, 16.98]

ans = df[df['new_price'].isin(lst)]

display(ans.head())
print(len(ans))
```
