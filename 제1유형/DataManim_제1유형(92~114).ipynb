# 비행탑승 경험 만족도 데이터

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

데이터 로드

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv")

print(df.shape)
df.head()
```

92. Arrival Delay in Minutes 컬럼이 결측치인 데이터들 중 'neutral or dissatisfied' 보다 'satisfied'의 수가 더 높은 Class는 어디 인가?

```python
target = df[df['Arrival Delay in Minutes'].isna()].groupby(['Class', 'satisfaction'], as_index=False).size().pivot(index='Class', columns='satisfaction')

ans = target[target['size']['neutral or dissatisfied'] < target['size']['satisfied']]
ans
```

# 의료 비용 예측 데이터

데이터 로드

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv")

print(df.shape)
df.head()
```

94. 흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 평균의 차이는?

```python
charges_yes = df[df['smoker'] == 'yes']['charges'].quantile(0.9)
charges_no = df[df['smoker'] == 'no']['charges'].quantile(0.9)

mean_yes = df[(df['smoker'] == 'yes') & (df['charges'] >= charges_yes)]['charges'].mean()
mean_no = df[(df['smoker'] == 'no') & (df['charges'] >= charges_no)]['charges'].mean()

ans = abs(mean_yes - mean_no)
ans
```

95. bedrooms 의 빈도가 가장 높은 값을 가지는 데이터들의 price의 상위 10%와 하위 10%값의 차이를 구하여라

```python
low = df[df['bedrooms'] == df['bedrooms'].value_counts().index[0]]['price'].quantile(0.1)
high = df[df['bedrooms'] == df['bedrooms'].value_counts().index[0]]['price'].quantile(0.9)

ans = high - low
ans
```

# 대학원 입학가능성 데이터

데이터 로드

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/train.csv")

print(df.shape)
df.head()
```

96. Serial No. 컬럼을 제외하고 'Chance of Admit'을 종속변수, 나머지 변수를 독립변수라 할때, 랜덤포레스트를 통해 회귀 예측을 할 떄 변수중요도 값을 출력하라 (시드값에 따라 순서는 달라질수 있음)

```python
df.drop(columns='Serial No.', inplace=True)

X = df.drop(columns='Chance of Admit')
y = df['Chance of Admit']


from sklearn.ensemble import RandomForestRegressor
from pandas import Series

# 객체 생성
ml = RandomForestRegressor()

# 모델 학습
ml.fit(X, y)
```

```python
feature_importance = ml.feature_importances_
Series_feat_imp = Series(feature_importance, index=X.columns)

ans = Series_feat_imp.sort_values(ascending=False)
ans
```

98. 남성들의 연령대별 (10살씩 구분 0~9세 10~19세 ...) Na_to_K값의 평균값을 구해서 데이터 프레임으로 표현하여라

# 센서데이터 동작유형 분류 데이터

데이터 로드

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/train.csv")

print(df.shape)
df.head()
```

100. pose값에 따른 각 motion컬럼의 중간값의 가장 큰 차이를 보이는 motion컬럼은 어디이며 그값은?

```python
target = df.groupby('pose').median().T

target['del'] = abs(target[0] - target[1])

ans = target[target['del'] == target['del'].max()].reset_index()['index'].values
ans
```
