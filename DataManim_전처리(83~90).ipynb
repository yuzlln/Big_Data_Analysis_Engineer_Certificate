# 6. Pivot

```python
import pandas as pd
import numpy as np
```

데이터 로드

```python
DataURL = 'https://raw.githubusercontent.com/Datamanim/pandas/main/under5MortalityRate.csv'
df = pd.read_csv(DataURL)

print(df.shape)
df.head()
```

83. Indicator을 삭제하고 First Tooltip 컬럼에서 신뢰구간에 해당하는 표현을 지워라

```python
# Indicator열 삭제
df.drop(columns='Indicator', inplace=True)

# 신뢰구간 표현 삭제
df['First Tooltip'] = df['First Tooltip'].map(lambda x : float(x.split('[')[0]))

ans = df.copy()
ans.head(2)
```

84. 년도가 2015년 이상, Dim1이 Both sexes인 케이스만 추출하라

```python
ans = df[(df['Period'] >= 2015) & (df['Dim1'] == 'Both sexes')].copy()
ans.head()
```

85. 84번 문제에서 추출한 데이터로 아래와 같이 나라에 따른 년도별 사망률을 데이터 프레임화 하라

```python
# pivot : 데이터가 고유한 값을 가져야 함
ans = ans.pivot(index='Location', columns='Period', values='First Tooltip')
ans.head()
```

86. Dim1에 따른 년도별 사망비율의 평균을 구하라

```python
# pivot_table : 동일한 인덱스, 열, 값 조합이 여러 개일 경우 집계 함수(기본값은 평균)를 적용
ans = df.pivot_table(index='Dim1', columns='Period', values='First Tooltip', aggfunc='mean')
ans.iloc[:, :4]
```

데이터 로드

```python
DataURL = 'https://raw.githubusercontent.com/Datamanim/pandas/main/winter.csv'
df = pd.read_csv(DataURL)

print(df.shape)
df.head()
```

87. 데이터에서 한국 KOR 데이터만 추출하라

```python
ans = df[df['Country'] == 'KOR']
ans.head()
```

88. 한국 올림픽 메달리스트 데이터에서 년도에 따른 medal 갯수를 데이터프레임화 하라

```python
kr = df[df['Country'] == 'KOR'].copy()

ans = kr.pivot_table(index='Year', columns='Medal', aggfunc='size').fillna(0)
ans
```

89. 전체 데이터에서 sport종류에 따른 성별수를 구하여라

```python
ans = df.pivot_table(index='Sport', columns='Gender', aggfunc='size')
ans
```

90. 전체 데이터에서 Discipline종류에 따른 따른 Medal수를 구하여라

```python
ans = df.pivot_table(index='Discipline', columns='Medal', aggfunc='size')
ans
```
